```python
def getListofsites(soup):
    website = {}
    div_wrapper = soup.body.div.main.div.find_all("div")
    links_soup = div_wrapper[4].div.ul.find_all("li")
    for link in links_soup:
        a_tag = link.a
        address = a_tag.attrs['href']
        name = a_tag.contents[0]
        website[name] = address
    
    return website
    
```

```python
def getAlphaOrderlist(soup):
    website = {}
    div_wrapper = soup.body.div.main.div.div.find_all("div")
    links_soup = div_wrapper[1].div.ol.find_all("li")
    for link in links_soup:
        a_tag = link.a
        address = a_tag.attrs['href']
        name = a_tag.contents[0]
        website[name] = address
        
    return website
```

```python
def getcompanyinfo(browser):
    # click on about
    #browser.find_element_by_xpath("//*[@class='org-page-navigation__items ']//*[text()='About']").click()
    #print(browser.page_source)   
    try:
        dl_tag = browser.find_element_by_xpath("//*[@class='overflow-hidden']")
    
        data_soup = bs.BeautifulSoup(dl_tag.get_attribute('innerHTML'))
        #print("done")
        #url scrapping
        url = data_soup.body.find_all("dd")[0].a.span.contents[0]

        #size of company scrapping
        size = data_soup.body.find_all("dd")[2].contents[0]
        size = size.strip(" ").strip("\n").strip(" ")

        name_sel = browser.find_element_by_xpath("//*[@class='org-top-card-summary__title t-24 t-black truncate']")
        name_soup = bs.BeautifulSoup(name_sel.get_attribute('innerHTML'))
        name = name_soup.body.span.contents[0]
        return [name,url,size]
    except:
        return ["None" , "none" , "NONE"]


```

```python
soup = bs.BeautifulSoup(browser.page_source)
sites = getListofsites(soup)
```

```python
Main_Data = []

for site in sites:
    browser.get(sites[site][:-20] + "/about/")
    print(len(Main_Data) + 1 , site)
    info = getcompanyinfo(browser)
    Main_Data.append(info)
    if(len(Main_Data)%50==0):
        input("enter anything")
    
    
```
